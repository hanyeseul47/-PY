```python
# 5명의 성적을 입력받아 평균을 내는 프로그램을 만든다
# 반복문과 조건문을 사용한다
# 변수는 total, counter, grade, averge
total = 0           #변수 total은 0으로 선언하였다
counter = 1             #변수 counter는 1로 선언
while counter <= 10  :          #반복문 p를 사용하여 counter가 10보다 작거나 같을때 다음 내부 문장으로 넘어간다
    grade = int (input("성적을 입력하시오.:"))  #변수 grade에 int(정수형 변환)와 사용자로부터 입력을 받는 input을 사용하여 정수를 받는다
    total = grade + total           #변수 total에 성적 grade와 total을 합한 값을 선언한다
    counter = counter + 1           #변수 counter는 초기 선언한 counter의 값에 실행할 때마다 1씩 추가된다
averge = total / 10             #변수 averge에서 계산한 토탈값을 10으로 나눈다
print(averge)           #평균값을 출력
#write by 한예슬
```

주어진 코드에서는 반복문과 조건문을 사용하여 성적을 입력받고, 입력받은 성적을 합산한 후 평균을 계산하여 출력합니다.

먼저, total 변수를 0으로 초기화하고, counter 변수를 1로 초기화합니다. 그리고 while 반복문을 사용하여 counter가 10보다 작거나 같을 때까지 아래의 내부 문장들을 반복 실행합니다.

반복문 내부에서는 grade 변수에 사용자로부터 성적을 입력받습니다. input() 함수를 사용하여 "성적을 입력하시오.:"라는 메시지를 출력하고, 입력된 성적을 int() 함수를 사용하여 정수로 변환합니다.

그리고 입력받은 성적을 total 변수에 더해줍니다. 이를 통해 입력받은 모든 성적의 합을 계산할 수 있습니다.

반복문이 한 번 실행될 때마다 counter 변수를 1씩 증가시킵니다. 이를 통해 몇 번째 성적인지를 나타내는 카운터 역할을 합니다.

반복문이 종료되면, total 변수에는 입력받은 성적들의 총합이 저장되어 있습니다. 이를 10으로 나누어 평균을 계산하고, average 변수에 저장합니다.

마지막으로, average 변수에 저장된 평균 값을 출력합니다.

따라서 위 코드를 실행하면 5명의 성적을 입력받아 평균을 계산하여 출력합니다.
